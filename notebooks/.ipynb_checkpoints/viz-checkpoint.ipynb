```lua
function scale_weights(weights, x_times)
	-- torch.setdefaulttensortype('torch.FloatTensor')
	ws	=	weights:float()
	w	=   torch.FloatTensor
							(
									(#ws)[2]
								,	(#ws)[3]*x_times
								,	(#ws)[4]*x_times
							)

	print( #ws )

	for	i	=	1,	(#ws)[1]	do
		for	j	=	1,	(#ws)[2]	do


			for	x	=	1,	(#ws)[3]*x_times		do
			for	y	=	1,	(#ws)[4]*x_times		do
				w[j][ x ][ y ] =
								ws[i][j]
										[ math.ceil(x/x_times) ]
										[ math.ceil(y/x_times) ]
				end
				end


			end
			itorch.image( w )
		end

	end




nodes, container_nodes = model:findModules('cudnn.SpatialConvolution')
print(#nodes)
```

```lua
 for i = 1, #nodes do
 	print(#nodes[i].weight)
    nodes[i].weight = nodes[i].weight:pow(3)
 	end
```
